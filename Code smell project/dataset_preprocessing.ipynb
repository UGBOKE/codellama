```python
import pandas as pd
import os


base_path_secure = r'C:\Users\UGBOKE GEORGE\Videos\Transformer\codesmell\codellama\Code smell project\Secure_SourceCodes\Secure_SourceCodes'
base_path_vulnerable = r'C:\Users\UGBOKE GEORGE\Videos\Transformer\codesmell\codellama\Code smell project\Vulnerable_SourceCodes\Vulnerable_SourceCodes'


def read_file_content(row):
 
    if row['label'] == 0:
        file_path = os.path.join(base_path_secure, f"{row['hash_id']}.sol")
    else:
        file_path = os.path.join(base_path_vulnerable, f"{row['hash_id']}.sol")
    

    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            return file.read()
    except Exception as e:

        return f"Error reading file: {str(e)}"


df_sorted['code'] = df_sorted.apply(read_file_content, axis=1)


print(df_sorted[['hash_id', 'label', 'code']])
```

```python
df1= df_sorted[['index','hash_id', 'label', 'code']]
```

```python
import pandas as pd
import re



def extract_comments(code):
    # Extract both single line and block comments
    single_line_comments = re.findall(r'//.*', code)
    block_comments = re.findall(r'/\*[^*]*\*+(?:[^/*][^*]*\*+)*/', code)
    # Combine all comments into one list
    all_comments = single_line_comments + block_comments
    return ' '.join(all_comments)

def remove_comments(code):
    # Remove single line comments
    code = re.sub(r'//.*', '', code)
    # Remove block comments
    code = re.sub(r'/\*[^*]*\*+(?:[^/*][^*]*\*+)*/', '', code)
    return code



# Extract comments and cleaned code
df1['comments'] = df1['code'].apply(extract_comments)
df1['smart_contracts'] = df1['code'].apply(remove_comments)

# Create a new DataFrame df2 with desired columns
df2 = df1[['label', 'smart_contracts', 'comments', 'index', 'hash_id']]

# Show the new DataFrame
print(df2.head())
```
